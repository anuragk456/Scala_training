```scala
import scala.io.Source
import scala.util.{Try, Success, Failure}

case class Employee(sno: Int, name: String, city: String, salary: Int, department: String) {
  override def toString: String = {
    s"Employee(SNo: $sno, Name: $name, City: $city, Salary: $$salary, Department: $department)"
  }
}

val file = "employee.csv";

val employees: List[Employee] = Try {
  Source.fromFile(file)
    .getLines()
    .drop(1) // Skip the header
    .map { line =>
      val Array(sno, name, city, salary, department) = line.split(",").map(_.trim)
      Employee(sno.toInt, name, city, salary.toInt, department)
    }.toList
} match {
  case Success(empList) => empList
  case Failure(exception) =>
    println(s"Error reading file: ${exception.getMessage}")
    List.empty[Employee] // Return an empty list on failure
}

// for(x<-employees){
//     println(x);
// }

val filtBySalary = employees.filter(_.salary > 60000)
// for(x<-filtBySalary){
//     println(x);
// }

val filtByDpt = employees.filter(_.department == "Engineering")
// for(x<-filtByDpt){
//     println(x);
// }

val res = employees.map(emp => s"${emp.name} from ${emp.city} earns ${emp.salary} in ${emp.department}")
var i = 0
// res.foreach { key =>
//   i += 1
//   println(s"$i: $key")
// }

// val res = employees.map(emp => emp.name -> s"from ${emp.city} earns ${emp.salary} in ${emp.department}").toMap;
// res.foreach { case (key, value) => println(s"Key: $key, Value: $value")};

val total = employees.map(_.salary).sum;
// println(s"total: ${total}");
val avg = if (employees.nonEmpty) total.toDouble / employees.size else 0.0;
// println(s"avg : ${avg}");

val dptWise = employees.groupBy(_.department).map {
  case (department, empList) => (department, empList.size)
}
var j = 0;
for(x<-dptWise){
    j+=1;
    println(s"${i}. ${x}");
}

```
